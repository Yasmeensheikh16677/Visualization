```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
import numpy as np
import pandas as pd
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
! pip install plotly
import plotly.graph_objects as go
import plotly.express as px
fig=px.scatter_3d(x=df.X, y=df.Y, z=df.Z)
fig.show()
```

```python
#2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
fig=px.violin(x=df.Grade, y=df.Score, )
fig.update_layout(
    title='Distribution of Scores across Different Grade Categories',
    xaxis=dict(title='Grade'),
    yaxis=dict(title='Score')
)
fig.show()
```

```python
#3.  Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
fig=px.imshow(df)
fig.update_layout(title='Heatmap chart for sales across month and days')
fig.show()
```

```python
fig.add_trace(go.Scatter(x=df['GDP'],y=df['Population'],mode='markers'))
fig.update_layout(title='Population vs GDP by Country',
    xaxis=dict(title='GDP'),
    yaxis=dict(title='Population'),
    showlegend=False)

# Show the chart
fig.show()
```
